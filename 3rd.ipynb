1] Used statistical functions on array min , max and median

```python
lst=[10,20,30,40,50]
print(f"List = {lst}")

import statistics as st
import numpy as np

arr=np.array(lst)
print(f"Array :- {arr}")
print()

print("Statistical Functions on array")

minimum=min(arr)
print(f"Minimum number from array = {minimum}")

maximum=max(arr)
print(f"maximum number from array = {maximum}")

median=np.median(arr)
print(f"median number from array = {median}")

mean=np.mean(arr)
print(f"mean number from array = {mean}")

standard_deviation = st.stdev(arr)
print(f"Statndard deviation of array = {standard_deviation}")
```

2] Calculating similarity and dissimilarity on given dataset using statistic methods.

```python
# 1] Using linalg.norm() method
from numpy.linalg import norm
point1 = np.array((4,4,2))
point2=np.array((1,2,1))
dist = np.linalg.norm(point1-point2)
print(dist)
```

```python
# 2] Using Dot product and square root functions
temp=point1-point2
dist=np.sqrt(np.dot(temp.T,temp))
print(dist)
```

```python
# 3] Using distance.cuclidian() from SciPy module

from scipy.spatial import distance
point1=(4,4,2)
point2=(1,2,1)
print(distance.euclidean(point1,point2))
```
